```
for col in ["person_age","person_emp_exp","loan_int_rate","cb_person_cred_hist_length"]:
 Q1 = df[col].quantile(0.25)
 Q3 = df[col].quantile(0.75)
 IQR = Q3 - Q1

 lower = Q1 - 1.5 * IQR
 upper = Q3 + 1.5 * IQR

 print(f"{col}: Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}")
 print(f"Lower bound: {lower}, Upper bound: {upper}")

 df[col] = df[col].clip(lower, upper)
 df['credit_score'] = df['credit_score'].clip(lower=300, upper=850)
```

```
df.plot(kind='box',subplots=True, figsize=(15, 10), color="#578BEC",layout=(5,5))
```

```
df['debt_to_income_ratio'] = df['loan_amnt'] / (df['person_income'] + 1)
```

```
df['loan_credit_ratio'] = df['loan_amnt'] / (df['credit_score'] + 1)
```

```
df['emp_stability'] = df['person_emp_exp'] / (df['person_age'] + 1)
##         ##
```

```
df['age_bucket'] = pd.cut(df['person_age'], bins=[18,25,35,50,65,100], labels=False)
```

```
df.drop(columns=['person_age','person_emp_exp'], inplace=True)
```

***Visualization***

```
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

n_cols = 3
n_rows = (len(cat_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(x=col, data=df, palette="Set2", ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
    axes[i].tick_params(axis="x", rotation=30)

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

df[numerical_cols].hist(bins=30, figsize=(15,10), color="skyblue", edgecolor="black")
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```
target = "loan_status"
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()

n_cols = 3
n_rows = -(-len(cat_cols) // n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5*n_rows))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(x=col, hue=target, data=df, palette="Set1", ax=axes[i])
    axes[i].set_title(f"{col} vs {target}")
    axes[i].tick_params(axis="x", rotation=30)

for j in range(len(cat_cols), len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```
num_cols = df.select_dtypes(include=["int64","float64"]).columns.tolist()
if target in num_cols: num_cols.remove(target)

n_cols = 3
n_rows = -(-len(num_cols) // n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5*n_rows))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(x=target, y=col, data=df, palette="Set1", ax=axes[i])
    axes[i].set_title(f"{col} vs {target}")

for j in range(len(num_cols), len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```
